```python
import requests
from bs4 import BeautifulSoup
```

```python
site_map="https://www.pngmart.com/sitemapindex.xml/"
```

```python
response=requests.get(site_map)
```

```python
response
```

```python
xml=response.text
```

```python
soup=BeautifulSoup(xml,"xml")
```

```python
soup
```

```python
sitemaps=[]
for i in soup.find_all('loc'):
    url=i.get_text()
    if "post-sitemap" in url : 
        sitemaps.append(i.get_text())
```

```python
sitemaps
```

```python
site_mmap_1=sitemaps[0]
site_mmap_1
```

```python
response=requests.get(site_mmap_1)
```

```python
response.text
```

```python
soup=BeautifulSoup(response.text,"xml")
soup
```

```python
master_list=[]
for i in soup.find_all('loc'):
    url=i.get_text()
    
    master_list.append(url)
```

```python
master_list[0:5]
```

```python
for image_url in master_list : 
    print(image_url)
    response=requests.get(image_url)
    
    break 
```
